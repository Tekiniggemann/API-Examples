# 2nd API

```python
resp = requests.get('https://www.fruityvice.com/api/fruit/all')
resp = resp.json()
resp
```

```python
fruit = pd.DataFrame(resp)
fruit.head(10)
```

```python
def flatten(fruit, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(fruit[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        fruit = pd.concat([fruit, flattened], axis=1)
        fruit = fruit.drop(column, axis=1)
    return fruit
```

```python
dict(fruit['nutritions'])
pd.DataFrame(dict(fruit['nutritions']))
pd.DataFrame(dict(fruit['nutritions'])).transpose()
```

```python
nested_columns = ['nutritions']

flat = flatten(fruit, nested_columns)
flat
```
